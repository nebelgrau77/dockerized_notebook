### Import necessary modules

```python
import matplotlib.pyplot as plt
```

```python
from sklearn import datasets
```

```python
iris = datasets.load_iris()
```

### Explore the dataset

```python
iris.feature_names
```

```python
iris.target_names
```

```python
X = iris.data[:, :2] # only the first two features
```

```python
y = iris.target
```

### Plot two selected features

##### Define some parameters for the chart

```python
chart_size = (10,8)
datapoint_size = 80 
color_palette = plt.cm.brg
datapoint_edge = 'k' # black edges of the data points
class_labels = iris.target_names
```

##### Set up the chart legend

```python
# map the y values to a 0,1 interval
norm = plt.Normalize(y.min(), y.max())

# define the legend symbols
handles = [plt.Line2D([0,0], [0,0], 
                      color = color_palette(norm(i)), 
                      marker = 'o', 
                      linestyle = '',
                      label = label) 
            for i, label in enumerate(class_labels)]
```

```python
# get objects from subplots, define chart size
fig, ax = plt.subplots(figsize = chart_size)

# draw a scatterplot 
ax.scatter(X[:,0], X[:, 1], c = y, s = datapoint_size, cmap=color_palette, edgecolor = datapoint_edge, norm=norm)

# add axis labels and the legend
ax.set_xlabel(iris.feature_names[0])
ax.set_ylabel(iris.feature_names[1])

ax.legend(handles=handles)
```
